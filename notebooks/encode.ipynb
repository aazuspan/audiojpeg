```python
from PIL import Image
from IPython.display import Audio
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import sys
from pathlib import Path

sys.path.insert(1, Path().resolve().parent.as_posix())

from audiojpeg import encoding
```

## Audio to Image

Load and preview the audio.

```python
path = "../data/PinkPanther30.wav"
sample_rate, audio = wavfile.read(path)
Audio(audio, rate=sample_rate)
```

Encode it into an image.

```python
encoded = encoding.encode_wav_to_image(path, order="F", width=512)
Image.fromarray(encoded[1:, ...]).save("../output/PinkPanther30_F.jpeg")
```

Visualize metadata encoding for Fortran order.

```python
fig, ax = plt.subplots(1, 1)

ax.imshow(np.expand_dims(np.array([1, 0, 0, 0, 1, 1, 0]), 0), cmap="gray")

ax.yaxis.set_ticks([])
ax.set_facecolor("black")
plt.margins(0)

plt.savefig("../output/order_encoding.png", bbox_inches='tight', transparent=True)
```
